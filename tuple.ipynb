Tuple is a sequence datatype which contains heterogeneous collection of elements and it is immutable. 

The inbuilt methods of tuple datatype are : 
1. len() - This method is used to find the length of the tuple.
2. count() - This method is used to count the number of occurences of an element in the tuple.
3. index() - This method is used to find the index of any element in the tuple. 



```python
# Defining a tuple : 
tuple1 = (1,2,3,4,5)
lst = [2,3,4]
tuple2 = tuple(lst)

# Define an empty tuple
tuple3 = tuple()

# Define a tuple with a single element
tuple4 = (3,)
print(tuple4)
print(type(tuple4))
```

```python
# Accessing an element from a tuple : 
tuple = (1,2,3,4,5,6,7,8)
tuple[3]

tuple1 = (1,2,3,"orange",['1','2','a',True,[100,200,300]],('a','b','c'))
tuple1[5] = 2344
tuple1
```

```python
# concatenation : 
tuple1 = (1,2,3)
tuple2 = (4,5,6)
tuple3 = (23,45,11,42)
result = tuple1 + tuple2 + tuple3
result
```

```python
# multiplication : 
tuple1 = (1,2,3)
result = tuple1 * 3
result
```

# Write a program to find the numbers between 100 and 400 where each digit of a number is even. Print the result in comma seperated values. 
(100,101,102,103,104... 400)


```python
number = 100
string = str(number)
for i in string:
    if int(i) % 2 == 0:
        status = True
    else:
        status = False
        break

if status == True:
    print("All the digits of given number are even")

```

```python
for j in range(100,400):
    number = j
    string = str(number)
    for i in string:
        if int(i) % 2 == 0:
            status = True
        else:
            status = False
            break

    if status == True:
        print(number, end = " , ")

```
